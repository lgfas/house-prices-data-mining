```python
# Contar valores ausentes em cada coluna
missing_values = train_df.isnull().sum()
missing_values[missing_values > 0]  # Mostrar apenas colunas com valores ausentes
```

```python
# 1️⃣ Preenchendo colunas categóricas com "Nenhum" (indica ausência da característica)
cols_com_nenhum = [
    "Alley", "FireplaceQu", "PoolQC", "Fence", "MiscFeature",  # Características opcionais
    "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",  # Porão
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",  # Garagem
    "MasVnrType"  # Tipo de revestimento de alvenaria
]
train_df[cols_com_nenhum] = train_df[cols_com_nenhum].fillna("Nenhum")

# 2️⃣ Preenchendo colunas numéricas com 0 (indica que a característica não existe)
cols_com_zero = ["MasVnrArea", "GarageYrBlt"]
train_df[cols_com_zero] = train_df[cols_com_zero].fillna(0)

# 3️⃣ Preenchendo LotFrontage com a mediana do bairro
train_df["LotFrontage"] = train_df.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))

# 4️⃣ Preenchendo a única entrada faltante em Electrical com a moda
train_df["Electrical"] = train_df["Electrical"].fillna(train_df["Electrical"].mode()[0])

# 5️⃣ Verificando se ainda existem valores ausentes
faltantes = train_df.isnull().sum().sum()

print(f"Total de valores ausentes após o tratamento: {faltantes}")  # Deve retornar 0 se tudo foi preenchido corretamente
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir algumas variáveis relevantes para análise de outliers
variaveis_relevantes = ["LotArea", "GrLivArea", "TotalBsmtSF", "1stFlrSF", "SalePrice"]

# Criar boxplots para visualizar outliers
plt.figure(figsize=(15, 8))
for i, coluna in enumerate(variaveis_relevantes, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=train_df[coluna])
    plt.title(f"Boxplot de {coluna}")

plt.tight_layout()
plt.show()
```

```python
# Definir função para remover outliers pelo método IQR
def remover_outliers(df, colunas):
    for coluna in colunas:
        Q1 = df[coluna].quantile(0.25)  # Primeiro quartil
        Q3 = df[coluna].quantile(0.75)  # Terceiro quartil
        IQR = Q3 - Q1  # Intervalo Interquartil
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR

        # Filtrar os dados dentro dos limites
        df = df[(df[coluna] >= limite_inferior) & (df[coluna] <= limite_superior)]

    return df

# Aplicar remoção de outliers
colunas_outliers = ["LotArea", "GrLivArea", "TotalBsmtSF", "1stFlrSF", "SalePrice"]
train_df = remover_outliers(train_df, colunas_outliers)

# Verificando nova quantidade de dados
print(f"Total de dados após remoção de outliers: {train_df.shape[0]} linhas")
```
