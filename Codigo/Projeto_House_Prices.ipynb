```python
# Contar valores ausentes em cada coluna
missing_values = train_df.isnull().sum()
missing_values[missing_values > 0]  # Mostrar apenas colunas com valores ausentes
```

```python
# 1️⃣ Preenchendo colunas categóricas com "Nenhum" (indica ausência da característica)
cols_com_nenhum = [
    "Alley", "FireplaceQu", "PoolQC", "Fence", "MiscFeature",  # Características opcionais
    "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2",  # Porão
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",  # Garagem
    "MasVnrType"  # Tipo de revestimento de alvenaria
]
train_df[cols_com_nenhum] = train_df[cols_com_nenhum].fillna("Nenhum")

# 2️⃣ Preenchendo colunas numéricas com 0 (indica que a característica não existe)
cols_com_zero = ["MasVnrArea", "GarageYrBlt"]
train_df[cols_com_zero] = train_df[cols_com_zero].fillna(0)

# 3️⃣ Preenchendo LotFrontage com a mediana do bairro
train_df["LotFrontage"] = train_df.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))

# 4️⃣ Preenchendo a única entrada faltante em Electrical com a moda
train_df["Electrical"] = train_df["Electrical"].fillna(train_df["Electrical"].mode()[0])

# 5️⃣ Verificando se ainda existem valores ausentes
faltantes = train_df.isnull().sum().sum()

print(f"Total de valores ausentes após o tratamento: {faltantes}")  # Deve retornar 0 se tudo foi preenchido corretamente
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir algumas variáveis relevantes para análise de outliers
variaveis_relevantes = ["LotArea", "GrLivArea", "TotalBsmtSF", "1stFlrSF", "SalePrice"]

# Criar boxplots para visualizar outliers
plt.figure(figsize=(15, 8))
for i, coluna in enumerate(variaveis_relevantes, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=train_df[coluna])
    plt.title(f"Boxplot de {coluna}")

plt.tight_layout()
plt.show()
```

```python
# Definir função para remover outliers pelo método IQR
def remover_outliers(df, colunas):
    for coluna in colunas:
        Q1 = df[coluna].quantile(0.25)  # Primeiro quartil
        Q3 = df[coluna].quantile(0.75)  # Terceiro quartil
        IQR = Q3 - Q1  # Intervalo Interquartil
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR

        # Filtrar os dados dentro dos limites
        df = df[(df[coluna] >= limite_inferior) & (df[coluna] <= limite_superior)]

    return df

# Aplicar remoção de outliers
colunas_outliers = ["LotArea", "GrLivArea", "TotalBsmtSF", "1stFlrSF", "SalePrice"]
train_df = remover_outliers(train_df, colunas_outliers)

# Verificando nova quantidade de dados
print(f"Total de dados após remoção de outliers: {train_df.shape[0]} linhas")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Criar histograma da variável SalePrice
plt.figure(figsize=(10, 5))
sns.histplot(train_cleaned_df["SalePrice"], bins=30, kde=True)
plt.title("Distribuição de SalePrice")
plt.xlabel("Preço das Casas")
plt.ylabel("Frequência")
plt.show()

# Teste de normalidade (Shapiro-Wilk)
stat, p = stats.shapiro(train_cleaned_df["SalePrice"])
print(f"Shapiro-Wilk Teste: Estatística={stat:.3f}, p-valor={p:.3f}")

# Aplicar log-transform para verificar se melhora a distribuição
train_cleaned_df["SalePrice_log"] = np.log1p(train_cleaned_df["SalePrice"])

# Plotar distribuição após log-transform
plt.figure(figsize=(10, 5))
sns.histplot(train_cleaned_df["SalePrice_log"], bins=30, kde=True)
plt.title("Distribuição de SalePrice (Após Log Transform)")
plt.xlabel("Log(Preço das Casas)")
plt.ylabel("Frequência")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Carregar os dados transformados
train_transformed_df = pd.read_csv("data/processed/train_transformed.csv")

# Selecionar apenas colunas numéricas (excluindo SalePrice original para evitar vazamento de dados)
colunas_numericas = train_transformed_df.select_dtypes(include=["number"]).columns
colunas_numericas = colunas_numericas.drop(["SalePrice"])  # Excluir a original, usaremos a log-transform

# Inicializar o StandardScaler
scaler = StandardScaler()

# Aplicar normalização
train_transformed_df[colunas_numericas] = scaler.fit_transform(train_transformed_df[colunas_numericas])

# Salvar os dados normalizados
caminho_arquivo_norm = "data/processed/train_normalized.csv"
train_transformed_df.to_csv(caminho_arquivo_norm, index=False)

print(f"✅ Dados normalizados e salvos! Arquivo: {caminho_arquivo_norm}")
```

```python
# Carregar os dados normalizados
train_normalized_df = pd.read_csv("data/processed/train_normalized.csv")

# Identificar colunas categóricas
categoricas = train_normalized_df.select_dtypes(include=["object"]).columns
print("🔹 Colunas categóricas no dataset:", categoricas)

# Aplicar One-Hot Encoding para variáveis categóricas
train_encoded_df = pd.get_dummies(train_normalized_df, columns=categoricas, drop_first=True)

# Salvar os dados prontos para modelagem
caminho_arquivo_final = "data/processed/train_final.csv"
train_encoded_df.to_csv(caminho_arquivo_final, index=False)

print(f"✅ Engenharia de características concluída! Dados salvos em: {caminho_arquivo_final}")
```
